# Figure Illustrating the Different Types of Noise in the Real Data 

```python
# import relevant libraries 
from matplotlib import pyplot as plt 
import seaborn as sns 
from src.data.utils import read_ns5_file

sns.set_theme()

def plot_data(eye : int, ramp : int, channel : int, xlim : tuple = (None, None), ylim : tuple = (None, None)) -> None :
    # read the data
    _, data = read_ns5_file(f'../../data/raw/Ramp data/Eye {eye+1}/ramp_file{ramp+1}.ns5')
    save_name = f"../../../figures/eye{eye+1}_ramp{ramp+1}_channel{channel+1}.pdf"
    # plot the data
    plt.figure(figsize=(18, 3))
    plt.plot(data[:, channel], color = "darkblue")
    if ylim[0] is not None:
        plt.ylim(ylim)
    else:
        plt.ylim([-7500, 7500])
    if xlim[0] is not None:
        plt.xlim(xlim)
    plt.savefig(save_name)
    plt.show()
```

```python
# define configuration 
eye = 2 
ramp = 2 
channel = 1 
# plot the data
plot_data(eye, ramp, channel)
```
