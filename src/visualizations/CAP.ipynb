```python
# plot normal distribution 
x = np.linspace(-10, 10, 100)
std = 3.5
y = np.exp(-x**2/(2*std**2))/(std*np.sqrt(2*np.pi))

plt.figure(figsize=(8, 4))
# plot the normal distribution
plt.plot(x, y, linewidth=2, color = "darkblue")

# plot filled segments 
plt.fill_between(x[:21], y[:21], color='skyblue', alpha=0.7)
plt.fill_between(x[40:60], y[40:60], color='skyblue', alpha=0.7)
plt.fill_between(x[79:100], y[79:100], color='skyblue', alpha=0.7)

# add vertical lines
plt.vlines([-10, -6, -2, 2, 6, 10], 0, 1, color='grey', linestyle='--', alpha=0.6)

# add text 
plt.text(-9.5, 0.13, "Fastest\n Fibers", fontsize=15, color='k')
plt.text(-1.5, 0.13, "Average\n Fibers", fontsize=15, color='k')
plt.text(6.5, 0.13, "Slowest\n Fibers", fontsize=15, color='k')

# plot things 
plt.xticks([])
plt.yticks([])
plt.title("Fiber Response Time", fontsize=20, weight = 'bold', pad = 20)
plt.ylim([0, np.max(y) + 0.05])

for spine in plt.gca().spines.values():
    spine.set_visible(False)
# plt.savefig("../../../figures/CAP.pdf")
plt.show()
```
