# Overview Figure in the top of Method Chapter

```python
# import relevant packages 
import matplotlib.pyplot as plt
import numpy as np 
import pywt 

rcParams = {
    "font.family": "serif",  # use serif/main font for text elements
    'text.usetex': True,
    'font.size': 8,
    'axes.labelsize': 7,
    'axes.titlesize': 9,
    'xtick.labelsize': 6,
    'ytick.labelsize': 6,
    'legend.fontsize': 7,
    'axes.labelpad': 1,
    'axes.axisbelow': True,  # draw gridlines below other elements
    "pgf.rcfonts": False,    # don't setup fonts from rc parameters
    "pgf.preamble": "\n".join([
        r"\usepackage{url}",            # load additional packages
        r"\usepackage{amsmath,amssymb}",   # unicode math setup
        #  r"\setmainfont{DejaVu Serif}",  # serif font via preamble
    ])
}
plt.rcParams.update(rcParams)

figdir = "../../../figures/methods/"
def savefig(fig, name, width=6, height=4, format = "pdf"):
    # apply rcParams
    fig.set_size_inches(width, height)
    dpi = 1000 if format == "png" else 'figure'
    fig.savefig(figdir + name + "." + format, bbox_inches='tight', dpi = dpi)


```

### Preprocessing Box 

```python
path = "../data/noise_files_sim_data/pli.npy"
data = np.load(path)

fig = plt.figure(figsize = (8, 2))
plt.plot(data, color = "black", linewidth = 0.4)
plt.xlim([0, 1500])
plt.xticks([])
plt.yticks([])
plt.show()

savefig(fig, "overview_preprocess1", width=1.5, height=1, format = "png")

fig = plt.figure(figsize = (8, 2))
plt.plot(data * 0.3, color = "black", linewidth = 0.4)
plt.xlim([0, 1500])
plt.ylim([data.min(), data.max()])
plt.xticks([])
plt.yticks([])
plt.show()

savefig(fig, "overview_preprocess2", width=1.5, height=1, format = "png")
```

### Method Box 

```python
# make threshold plot 

x = np.zeros(100)
idx = np.arange(10, 20)
x[idx] = [1, 2, 3, 4, 5, 5, 4, 3, 2, 1]
x += np.random.normal(0, 0.1, 100)

fig = plt.figure(figsize = (8, 2))
plt.plot(np.arange(100), x, color = "black", linewidth = 0.6)
plt.plot(np.arange(100), np.ones(100) * 3.5, "--", color = "red", linewidth = 0.6)
plt.xticks([])
plt.yticks([])
plt.show()

savefig(fig, "overview_method1", width=1.5, height=0.5, format = "png")
```

```python
wavelet = pywt.ContinuousWavelet('morl')

fig, ax = plt.subplots(1, 1)
ax.plot(np.real(wavelet.wavefun()[0]), color = "black", linewidth = 0.6)
plt.xticks([])
plt.yticks([])
plt.show()

savefig(fig, "overview_method2", width=1.5, height=0.5, format = "png")
```

```python
x = np.linspace(-3, 8, 100)
fx = x**3 - 7*x**2   # cubic function
fxp = fx + 50
fxm = fx - 50

# Create the figure
fig, ax = plt.subplots(1, 1)
ax.plot(x, fx, color="black", linewidth = 0.6)
ax.plot(x, fxp, '--', color="black", linewidth = 0.6)
ax.plot(x, fxm, '--', color="black", linewidth = 0.6)

# add points in plot 
ax.plot([1, -2, 5], [15, -60, -40], 'o', color="black", markersize=1.5)
ax.plot([2, -1, 6], [50, -100, 75], 'o', color = "black", markersize=1.5)

plt.xticks([])
plt.yticks([])
plt.show()

savefig(fig, "overview_method3", width=1.5, height=0.5, format = "png")
```
